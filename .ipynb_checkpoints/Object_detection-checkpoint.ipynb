```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
yolo = cv2.dnn.readNet("yolov3.weights", "yolov3.cfg")
```

```python
classes = []
with open("coco.names","r") as file:
    classes = [line.strip() for line in file.readlines()]
```

```python
layer_names = yolo.getLayerNames()
outputlayers = [layer_names[i[0] -1] for i in yolo.getUnconnectedOutLayers()]
```

```python
colorRed   = (0, 0, 255)
colorGreen = (0, 255, 0)
```

```python
df = pd.DataFrame(columns=["filename", "no.car","no.person"])
```

```python
import sys
path="dataset/"    #  <--- change path here
filenames = next(os.walk(path))[2]
for f in filenames:
    #print(f)
    img = cv2.imread("dataset/"+f)
    #plt.imshow(img)
    height, width, channels = img.shape
    blob = cv2.dnn.blobFromImage(img, 1 / 255.0, (416, 416), swapRB=True, crop=False)
    yolo.setInput(blob)
    outputs = yolo.forward(outputlayers)
    class_ids = []
    confidences = []
    boxes = []
    for output in outputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if (confidence > 0.5):
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)
                x = int(center_x - w/2)
                y = int(center_y - h/2)
                
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)
        indices = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)
        colors = np.random.uniform(0, 255, size=(len(classes), 3))
        car=person=motor=traffic=bus=truck=0
        for i in range(len(boxes)):
            if i in indices:
                x, y, w, h = boxes[i]
                try:
                    #print("The entry is")
                    label =str(classes[class_ids[i]])
                except:
                    print("Oops!", sys.exc_info()[0], "occurred.")
                    break
                if label=="car":
                    car=car+1
                elif label=="person":
                    person=person+1
                
                
    df.loc[len(df.index)] = [f, car,person]
        
        
```

```python
person_result=df.loc[df['no.person'] != 0]
```
